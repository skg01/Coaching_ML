```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.stats as stats
import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = 20, 5
```

```python
# build gamma distribution as population
shape, scale = 2., 2.  # mean=4, std=2*sqrt(2)
p = np.random.gamma(shape, scale, 1000000)
b = plt.hist(p,200, density=True)
```

```python
# Step 2
## sample from population with different number of sampling size
# a list of sample mean
meansample = []
# number of sample
numofsample = 10000
# sample size
samplesize = [500,1000,10000,50000]
# for each number of sampling (1000 to 50000)
for i in samplesize:
    # collect mean of each sample
    eachmeansample = []
    # for each sampling
    for j in range(0,numofsample):
        # sampling 500 sample from population
        rc = random.choices(p, k=i)
        # collect mean of each sample
        eachmeansample.append(sum(rc)/len(rc))
    # add mean of each sampling to the list
    meansample.append(eachmeansample)
```

```python
# Step 3
cols = 2
rows = 2
sn = 0
for i in range(0, rows):
    for j in range(0, cols):
        plt.subplot(cols,rows,n+1)
        sns.distplot(meansample[n], bins=200)
        plt.title(label=" Sample Size :" + str(samplesize[n]))
        n += 1
plt.show()
```
